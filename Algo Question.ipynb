# Algorithmic Question (AQ)

#### *Python Implementation*

```python
from typing import List
```

```python
def check_sum(test_case: int, n: List[int], k: List[int]):

    for i in range(test_case):

        # conditional check if the result can be achieved through odd parity
        if n[i] >= k[i] and (n[i] - k[i]) % 2 == 0:
            print("YES")
            print("1 " * (k[i] - 1) + str(n[i] - (k[i] - 1))) # outputting odd integers that add up to the given sum
            
        # conditional check if the result can be achieved through even parity
        elif n[i] >= 2 * k[i] and (n[i] - 2 * k[i]) % 2 == 0:
            print("YES")
            print("2 " * (k[i] - 1) + str(n[i] - 2 * (k[i] - 1))) # outputting even integers that add up to the given sum

        # if the result cannot be achieved
        else:
            print("NO")
```

```python
# inputting no. of test cases
print("INPUT")
t = int(input()) 

n = list()
k = list()

# inputting sum and no. of integers for each test case
for i in range(t):
    sum_val, no_int = map(int, input().split())
    n.append(sum_val)
    k.append(no_int)

print("\nOUTPUT")
check_sum(t, n, k)
```

#### *Time Complexity*

We will be looking at only the `check_sum` function to calculate the Big O notation for our python code, as going over the input procedure does not really translate to the actual time complexity of our implementation, because the core algorithm is always the focus.  
  
If we look at the function, we can see there is a for loop that runs for the no. of test cases i.e. t. Hence the time complexity for the loop will be **O(n)**. As for each of the conditional checks within the loop and the print statements, they are of constant time i.e. **O(1)**.  

However, a string of k[i] integers is also printed, once for each test case, which can amount to time complexity of **O(k[i])** 

Therefore, the total time complexity of our code in Big O notation is **O(n * k[i])**.
